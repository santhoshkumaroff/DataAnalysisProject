> Numpy


> Pandas

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

# **DATA EXTRACTION**
> By using pandas to read csv file from local storage




```python
df = pd.read_csv('sales_data.csv',encoding='ISO-8859-1')
```

```python
df.head()
```

```python
# Information of dataframe
df.info()
```

```python
# Removing Unwanted Column
df.drop(['Status','unnamed1'],axis=1,inplace=True)
df.head()
```

```python
# Removing NaN values From dataframe
df.dropna(inplace=True)
df.info()
```

```python
# replace values present in Gender 'F' --> 'Female','M' --> 'Male'
df['Gender']=df['Gender'].map({'F':'Female','M':'Male'})
df.head()
```

```python
# replace values present in Marital_Status 0 --> Unmarried,1 --> Married
df['Marital_Status']=df['Marital_Status'].map({0:'Unmarried',1:'Married'})
df.head()
```

```python
# Removing Unwanted Column = 'Age' based on our requirements
df.drop('Age',axis=1,inplace=True)
df.head()
```

# **Data Analysis Part**

# **Gender**

```python
a = sns.countplot(x='Gender',data=df)
for i in a.containers:
  a.bar_label(i)
```

```python
b = df.groupby('Gender')['Amount'].sum()
b = pd.DataFrame(b)
b.head()
```

```python
a = sns.barplot(x='Gender',y='Amount',data=b)
for i in a.containers:
  a.bar_label(i)
```

## **Conclusion :** Based on above analysis **`Females`** are spending more amount on shopping


---
